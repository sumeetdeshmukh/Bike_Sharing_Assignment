```python
#mapping of weatherlisr column
bike_dataset['weathersit'] = bike_dataset['weathersit'].map({1:'Clear',2:'Mist',3:'Light Rain',4:'Heavy rain'})
```

```python
#After performing cleaning and visualizing dataset actions  we can now starting models
#But first we will create dummy variables
```

```python
#Creating dummy variables for month, weekday, weathersit and season variables.

#the reason we are converting the datatypes of int type because there is change that these will be created boolean('True','False') as default
#With boolean datatype we might face issue while building models
bike_months=pd.get_dummies(bike_dataset['mnth'],drop_first=True)
bike_months = bike_months.astype(int)

bike_weekdays=pd.get_dummies(bike_dataset['weekday'],drop_first=True)
bike_weekdays = bike_weekdays.astype(int)

bike_weathersit=pd.get_dummies(bike_dataset['weathersit'],drop_first=True)
bike_weathersit = bike_weathersit.astype(int)

bike_seasons=pd.get_dummies(bike_dataset['season'],drop_first=True)
bike_seasons = bike_seasons.astype(int)
```

```python
#creating new dataset and merging the dummy variable in it,this way we can also preseve the dataset used earlier for further use
bike_new_dataset = pd.concat([bike_dataset,bike_months,bike_weekdays,bike_weathersit,bike_seasons],axis=1)
```

```python
#first 5 elements of newly created dataset
bike_new_dataset.head()
```

```python
#info about newly created dataset
bike_new_dataset.info()
```

```python
#dropping the columns from which we created dummy variables as they are now unneccasry as well as atemp which we was correlated with temp
bike_new_dataset.drop(['season','mnth','weekday','weathersit','atemp'], axis = 1, inplace = True)
```

```python
#verifying dataset
bike_new_dataset.head()
```

```python
#All is set now we can split the data into train and test
```

```python
#Using 65% data for train and 35% for test
#using random seed as 0 to obtain reproducible results across different runs.
np.random.seed(0)
bike_dataset_train, bike_dataset_test = train_test_split(bike_new_dataset, train_size = 0.65,test_size=0.35,random_state =200)
```

```python
#Now we check the size of test and train dataset
```

```python
bike_dataset_train.shape
```

```python
bike_dataset_test.shape
```

```python
#we are going to scale our training dataset
scaler = MinMaxScaler()
```

```python
# Applying  scaler() to all numerical variables 

num_vars = ['temp','hum','windspeed','cnt']
bike_dataset_train[num_vars] = scaler.fit_transform(bike_dataset_train[num_vars])
```

```python
#now checking the head of dataset
bike_dataset_train.head()
```

```python
## check the correlation coefficients to see which variables are highly correlated
plt.figure(figsize = (25,25))
sns.heatmap(bike_dataset_train.corr(), cmap='RdYlGn', annot = True,mask= np.triu(bike_dataset_train.corr()))
plt.show()
```

```python
#cnt seems to have correlation with year variable and temp
```

```python
# Building the Linear Model

#So here we are separating the target variable cnt from the features
y_train = bike_dataset_train.pop('cnt')
X_train = bike_dataset_train

# Creating linear regression model and fitting the model with input data X_train and target variable y_train
lm_model = LinearRegression()
lm_model.fit(X_train,y_train)


#We are using Recursive Feature Elimination (RFE) which is a a linear regression model to select the top 15 most important features from the dataset
rfe = RFE(lm_model)
rfe = rfe.fit(X_train, y_train)
```

```python
# selecting the selected variable via RFE
column_selected= X_train.columns[rfe.support_]
```

```python
# Function to calculate VIF of variables

def get_vif(ds):
    vif = pd.DataFrame()
    vif['Features'] = ds.columns
    vif['VIF'] = [variance_inflation_factor(ds.values, i) for i in range(ds.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
#Taking dataframe with RFE selected variables and and calculating vif
X_train_rfe = X_train[column_selected]
get_vif(X_train_rfe)
```

```python
#hum shows high VIF values 
```

```python
#1st linear model
X_train_lm_1 = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
print(lr_1.summary())
```

```python
# As hum shows high VIF values hence we can drop it
X_train_new = X_train_rfe.drop(['hum'], axis = 1)

# Run the function to calculate VIF for the new model
get_vif(X_train_new)
```

```python
# Building 2nd linear regression model

X_train_lm_2 = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

```python
# We can drop February variable as it has high p-value
X_train_new = X_train_new.drop(['February'], axis = 1)

# Run the function to calculate VIF for the new model
get_vif(X_train_new)
```

```python
# Building 3rd linear regression model

X_train_lm_3 = sm.add_constant(X_train_new)
lr_3 = sm.OLS(y_train,X_train_lm_3).fit()
print(lr_3.summary())
```

```python
# We can drop January variable as it has high p-value
X_train_new = X_train_new.drop(['January'], axis = 1)

# Run the function to calculate VIF for the new model
get_vif(X_train_new)
```

```python
# Building 4th linear regression model

X_train_lm_4 = sm.add_constant(X_train_new)
lr_4 = sm.OLS(y_train,X_train_lm_4).fit()
print(lr_4.summary())
```

```python
# We can drop july variable as it has high p-value,though holiday also has same p-value,we have chosen july
X_train_new = X_train_new.drop(['July'], axis = 1)

# Run the function to calculate VIF for the new model
get_vif(X_train_new)
```

```python
# Building 5th linear regression model

X_train_lm_5 = sm.add_constant(X_train_new)
lr_5 = sm.OLS(y_train,X_train_lm_5).fit()
print(lr_5.summary())
```

```python
#F-Statistics value of 226.1 (which is greater than 1) and the p-value  is  almost equals to zero, 
#states that the overall model is significant
#So we will consider lr_5 model
```
